```python
import ipywidgets as widgets
from IPython.display import display, clear_output

# Instruction
instruction = widgets.HTML("<b>Write the line of code to create a list called <code>nums</code> with values 1, 2, 3.</b>")

# Text input for user code
code_input = widgets.Textarea(
    value='',
    placeholder='Enter your code here...',
    description='Code:',
    layout=widgets.Layout(width='100%', height='60px')
)

# Buttons
submit_button = widgets.Button(description='Submit')
hint_button = widgets.Button(description='Show Hint')

# Output
output = widgets.Output()
hint_output = widgets.Output()

# Expected variables
expected = [1, 2, 3]

# Submit function
def evaluate_code(b):
    output.clear_output()
    hint_output.clear_output()
    user_code = code_input.value
    try:
        local_vars = {}
        exec(user_code, {}, local_vars)
        if 'nums' in local_vars and local_vars['nums'] == expected:
            with output:
                print("‚úÖ Correct! You've created the list properly.")
        else:
            with output:
                print("‚ùå Incorrect. Make sure you create a list with values 1, 2, 3 and call it 'nums'.")
    except Exception as e:
        with output:
            print("‚ö†Ô∏è Error in your code:", e)

# Hint function
def show_hint(b):
    hint_output.clear_output()
    with hint_output:
        print("üîç Hint: Use square brackets to create a list: nums = [1, 2, 3]")

# Bind buttons
submit_button.on_click(evaluate_code)
hint_button.on_click(show_hint)

# Display all
display(instruction, code_input, widgets.HBox([submit_button, hint_button]), output, hint_output)
```
